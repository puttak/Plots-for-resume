```python
# library
import matplotlib.pyplot as plt
from collections import OrderedDict
from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
#rc('text', usetex=True)
# create data
names='Creative', 'Passionate', 'Effective communication', 'Solution oriented','well-organized','Integrity','team player'
size=[14,14,14,14,14,15,15]
 
# Create a circle for the center of the plot
my_circle=plt.Circle( (0,0), 0.65, color='white')

# Give color names
plt.pie(size, labels=names, colors=['red','green','blue','indigo','teal','olive','yellowgreen'])
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import rc

# colors https://matplotlib.org/3.2.1/gallery/color/named_colors.html

# The slices will be ordered and plotted counter-clockwise.
labels = 'Creative', 'Passionate', 'Effective communication', 'Solution oriented','well-organized','Integrity','team player'
sizes = [15, 14, 14, 14,14, 14, 15]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','darkorchid','deeppink','turquoise']
explode = (0, 0, 0, 0,0,0,0)  # explode a slice if required

#plt.pie(sizes, explode=explode, labels=labels, colors=colors,
#        autopct='%1.1f%%', shadow=True)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,shadow=True)
        
#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=2.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.rc('text', usetex=True)
# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
plt.show() 
#plt.savefig('plot.pdf', dpi = 600)
```
